```python
def ATNF_data_frame():
    return query_ATNF().pandas
```

```python
df = ATNF_data_frame()
df
```

```python
def get(psrj_name: str):
    _df = ATNF_data_frame()
    df=_df[_df["PSRJ"] == psrj_name].reset_index(drop=True)
    return df.iloc[0]
```
